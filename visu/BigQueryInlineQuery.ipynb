```python
%%bigquery nox
SELECT
    TBTimestamp AS ts,
    NO_Scaled AS nox
FROM `exeter-science-unit.airmonitor.airmonitor`
WHERE UniqueId = 128150 
    AND NO_Status = 'Valid'
    AND NO_Scaled >= 0
    AND TBTimestamp > '2018-10-01 11:33:03.984385+00:00'
ORDER BY ts
```

```python
%%bigquery co
SELECT
    TBTimestamp AS ts,
    CO_Scaled AS co
FROM `exeter-science-unit.airmonitor.airmonitor`
WHERE UniqueId = 128150 
    AND CO_Status = 'Valid'
    AND CO_Scaled >= 0
    AND TBTimestamp > '2018-10-01 11:33:03.984385+00:00'
ORDER BY ts
```

```python
%%bigquery no2
SELECT
    TBTimestamp AS ts,
    NO2_Scaled AS no2
FROM `exeter-science-unit.airmonitor.airmonitor`
WHERE UniqueId = 128150 
    AND NO2_Status = 'Valid'
    AND NO2_Scaled >= 0
    AND TBTimestamp > '2018-10-01 11:33:03.984385+00:00'
ORDER BY ts
```

```python
# Setup 
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111)

# plotting of previously queried data
nox.plot(x='ts', y='nox', style='.', ms=3, ax=ax, label='$\mathsf{NO}$')  # ms is markersize
co.plot(x='ts', y='co', style='.', ms=3, ax=ax, label='$\mathsf{CO}$')
no2.plot(x='ts', y='no2', style='.', ms=3, ax=ax, label='$\mathsf{NO_2}$')

# cosmetics 
lines = ax.get_lines()  # needed for legend
ax.legend(lines, [l.get_label() for l in lines], loc=2, fontsize=12)  # recreate legend
ax.set_ylabel('Parts per Billion', fontsize=14)
ax.set_xlabel('Time', fontsize=14)
```
